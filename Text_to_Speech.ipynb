```python
from gtts import gTTS
import os

def text_to_speech(text):
    tts = gTTS(text=text, lang='en')
    tts.save("output.mp3")
    os.system("start output.mp3")

def chatbot():
    print("Hello! I'm your text-to-speech chatbot. Type 'exit' to stop.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break
        print(f"Chatbot: {user_input}")
        text_to_speech(user_input)

chatbot()
```

```python
# --- 1. Install once per session ---
!pip -q install gTTS

# --- 2. Helper function ---
from gtts import gTTS
from IPython.display import Audio, HTML
from google.colab import files

def speak(text, mode="widget", lang="en"):
    """
    Generate speech from text in Google Colab.

    Parameters
    ----------
    text : str
        The sentence(s) to speak.
    mode : str, optional
        "widget"  – Colab inline audio player (default)
        "html"    – HTML <audio> tag (streaming)
        "download"– Push file to your browser for download
    lang : str, optional
        Language code (e.g. 'en', 'hi', 'fr', 'es', ...)

    Returns
    -------
    IPython display object or None
    """
    fname = "speech.mp3"
    gTTS(text, lang=lang).save(fname)

    if mode == "widget":
        return Audio(fname, autoplay=True)
    elif mode == "html":
        return HTML(f'<audio controls autoplay src="{fname}"></audio>')
    elif mode == "download":
        files.download(fname)
        print("⬇️  Download started in your browser")
    else:
        raise ValueError("mode must be 'widget', 'html', or 'download'.")

# --- 3. Examples ---
# Inline player (default)
speak("Hi there! I am Avanthika from CubeAI Solutions.")
#speak("नमस्ते दुनिया", lang="hi")
#speak("நீங்கள் எப்படி இருக்கிறீர்கள்?")
# HTML streaming
# speak("Streaming with an HTML audio tag.", mode="html")

# Download file
# speak("You can download me and listen offline.", mode="download")
```
